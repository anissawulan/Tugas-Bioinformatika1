```python
#import modul
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
```

```python
#Sequences
X = "AACCCGGTCGTA"
Y = "AAATTAAGCCCGATA"
```

```python
#Parameter;
#match = +5, mismatch = -1, open dan extend gap = -0.1
alignments = pairwise2.align.globalms(X, Y, 5, -1, -0.1, -0.1)
```

```python
#Menggunakan format_allignment untuk menunjukkan alignment
for t in alignments:
    print(format_alignment(*t))
    
```
